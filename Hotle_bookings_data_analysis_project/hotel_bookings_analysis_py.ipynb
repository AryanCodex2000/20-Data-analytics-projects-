##Data Cleaning

###Fetching the duplicate values in the records if any

```python
df.duplicated().sum()
```

###Checking for the unique values in the record

```python
df.nunique()
```

```python
df.nunique().sort_values()
```

```python
# unique counts
df.nunique().sort_values()

# low-cardinality columns
low_unique = df.nunique()[df.nunique() <= 10].sort_values()

# high-cardinality columns
high_unique = df.nunique()[df.nunique() > 50].sort_values(ascending=False)
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
# company me 93% null hai — column drop karo, kisi kaam ka nahi
df.drop(columns=['company'], inplace=True)

# children me sirf 4 nulls hain — 0 se fill karo (mode = 0, matlab mostly no children)
df['children'] = df['children'].fillna(0)

# country me kuch nulls hain — "Unknown" se fill karo
df['country'] = df['country'].fillna("Unknown")

# agent me nulls hain — 0 se fill karo (no agent used)
df['agent'] = df['agent'].fillna(0)
```

```python
df.isnull().sum()
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
print(df.isnull().sum().sort_values(ascending=False))
```

```python
df.drop(columns=['name', 'email', 'phone-number', 'credit_card'], inplace=True)
```

```python
print("Final shape:", df.shape)
df.head()
```

#### Hotel Wise Bookings

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="hotel", palette="Set2")
plt.title("Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
plt.show()
```

#### Cancellation count

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="is_canceled", palette="Set3")
plt.title("Cancellation Status")
plt.xlabel("Is Canceled")
plt.ylabel("Count")
plt.show()
```

#### Hotel Vs Cancellation

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="hotel", hue="is_canceled", palette="Set3")
plt.title("Hotel Type Vs Cancellation")
plt.xlabel("Hotel Type")
plt.ylabel("Count")
plt.show()
```

#### Monthly Booking Trend

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="arrival_date_month", order=["January","February","March","April","May","June",
    "July","August","September","October","November","December"], palette="viridis")
plt.xticks(rotation=45)
plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.show()
```

#### Country Wise Top Guests

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(6,4))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='magma')
plt.title('Top 10 Countries with Most Guests')
plt.xlabel('Country')
plt.ylabel('Bookings')
plt.show()
```

#### Market Segment Analysis

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="market_segment", order=df["market_segment"].value_counts().index, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Bookings by market segmment")
plt.xlabel("Market Segment")
plt.ylabel("Count")
plt.show()
```

#### Lead Time Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="lead_time", bins=50, kde=True, color="teal")
plt.title("Lead Time Distribution")
plt.xlabel("Lead Time")
plt.ylabel("Frequency")
plt.show()
```

#### ADR Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df["adr"], bins=50, kde=True, color="orange")
plt.title("ADR Distribution")
plt.xlabel("ADR")
plt.ylabel("Frequency")
plt.show()
```

#### Hotel wise ADR

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x="hotel", y="adr", palette="Set2")
plt.title("ADR by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("ADR")
plt.show()
```

#### Lead Time vs Cancellation

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x="is_canceled", y="lead_time", palette="Set1")
plt.title("Lead Time vs Cancellation")
plt.xlabel("Is Canceled")
plt.ylabel("Lead Time")
plt.show()
```

#### Special Requests vs Cancellation

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="total_of_special_requests", hue="is_canceled", palette="Set2")
plt.title("Special Requests vs Cancellation")
plt.xlabel("Total Special Requests")
plt.ylabel("Count")
plt.show()
```

#### Correlation Heatmap

```python
num_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))
sns.heatmap(num_df.corr(), annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

#### Stay Length Analysis

```python
df["total_stay"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]

plt.figure(figsize=(8,4))
sns.histplot(df["total_stay"], bins=30, kde=True, color="purple")
plt.title("Total Stay Distribution")
plt.xlabel("Total Stay Nights")
plt.ylabel("Frequency")
plt.show()
```

#### Customer Type Analysis

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="customer_type", order=df["customer_type"].value_counts().index, palette="Set3")
plt.title("Bookings by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Count")
plt.show()
```

#### Repeated Guest vs Cancellation

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="is_repeated_guest", hue="is_canceled", palette="Set1")
plt.title("Repeated Guest vs Cancellation")
plt.xlabel("Is Repeated Guest")
plt.ylabel("Count")
plt.show()
```
