```python
#Let's find out the maximum users daily engaged
df.columns = df.columns.str.strip()

# Convert the 'Daily_Engaged_Users' column to a numeric type. This is necessary
# in case the column was loaded as an object (string) due to the file structure.
# The 'coerce' option will turn any non-numeric values into NaN.
df['Daily_Engaged_Users'] = pd.to_numeric(df['Daily_Engaged_Users'], errors='coerce')

# Find the maximum value in the 'Daily_Engaged_Users' column.
max_interaction = df['Daily_Engaged_Users'].max()
#for minimum
min_interaction = df['Daily_Engaged_Users'].min()

# Find the row with the maximum interaction to get the corresponding date.
# We use idxmax() to find the index of the maximum value.
date_of_max_interaction = df.loc[df['Daily_Engaged_Users'].idxmax()]['Date']
date_of_min_interaction = df.loc[df['Daily_Engaged_Users'].idxmax()]['Date']

# Print the results in a clear format.
print(f"The maximum daily interaction of users is: {max_interaction}")
print(f"This maximum occurred on: {date_of_max_interaction}")
print(f"The minimum daily interaction of users is: {min_interaction}")
print(f"The minimum occurred on: {date_of_min_interaction}")
```

```python
def find_max_values(file_path):
    try:
        # Load dataset
        df = pd.read_csv('/kaggle/input/socialmedia-dataset-for-data-analysis/social_media.csv', header=1)

        # Clean column names
        df.columns = df.columns.str.strip()  # Remove accidental spaces

        # Rename first column to 'Date'
        df.rename(columns={'Unnamed: 0': 'Date'}, inplace=True)

        # Convert Date column
        df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')

        # Drop rows with invalid dates
        if df['Date'].isnull().any():
            print("Warning: Some dates could not be parsed and were dropped.")
            df = df.dropna(subset=['Date'])  # safer: assign back instead of inplace

        if df.empty:
            print("Error: No valid data after cleaning Date column.")
            return

        # Validate key columns exist
        required_cols = ['Daily_Engaged_Users', 'Daily_Total_Reach']
        for col in required_cols:
            if col not in df.columns:
                print(f"Error: Required column '{col}' not found. Available columns: {list(df.columns)}")
                return

        # Check for all-NaN in key columns
        if df['Daily_Engaged_Users'].isna().all():
            print("Error: 'Daily_Engaged_Users' has no valid data.")
            return
        if df['Daily_Total_Reach'].isna().all():
            print("Error: 'Daily_Total_Reach' has no valid data.")
            return

        # Find max rows
        max_engaged_row = df.loc[df['Daily_Engaged_Users'].idxmax()]
        max_reach_row = df.loc[df['Daily_Total_Reach'].idxmax()]

        # Safe date formatting
        def safe_date_str(date_val):
            return date_val.strftime('%Y-%m-%d') if pd.notna(date_val) else "Unknown Date"

        # Print results
        print("Analysis of Daily Engagement and Reach:\n")
        print("Maximum Daily Engaged Users:")
        print(f"  Value: {max_engaged_row['Daily_Engaged_Users']}")
        print(f"  Date:  {safe_date_str(max_engaged_row['Date'])}\n")

        print("Maximum Daily Total Reach:")
        print(f"  Value: {max_reach_row['Daily_Total_Reach']}")
        print(f"  Date:  {safe_date_str(max_reach_row['Date'])}")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
